# Line Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#To plot a line
x= np.arange(100)
y=x**2+10
plt.plot(x,y)
```

```python
#Line Plot
x = np.arange(10)
y1 = x**2
y2 = 2*x+3
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
#Themes available
thems = plt.style.available
print(thems)
```

```python
plt.style.use('seaborn')
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='blue')

plt.show()
```

```python
plt.style.use('seaborn')
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='blue')
#this output wont be shown outside jupyter notebook because its a object
#its ability of jupyter to show it. So we need to use plt.show()
#plt.show() also indicates that graph is complete
```

```python
plt.style.use('seaborn')
plt.plot(x,y1,color='red')
plt.show()
plt.plot(x,y2,color='blue')
plt.show()
```

```python
#To give titles
plt.style.use('seaborn')
plt.plot(x,y1,color='red')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Fruits")
plt.show()
```

```python
#To show color indication
plt.style.use('seaborn')
plt.plot(x,y1,color='red',label="Kiwi")
plt.plot(x,y2,color='blue',label="Mango")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Fruits")
plt.legend()
plt.show()
```

### All function to see documentation
#### plt.plot?

```python
#To show color indication
plt.style.use('seaborn')
plt.plot(x,y1,color='red',label="Kiwi",linestyle="dashed")
plt.plot(x,y2,color='blue',label="Mango")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Fruits")
plt.legend()
plt.show()
```

```python
#To show color indication
plt.style.use('seaborn')
plt.plot(x,y1,color='red',label="Kiwi",marker='o')
plt.plot(x,y2,color='blue',label="Mango",marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Fruits")
plt.legend()
plt.show()
```
